```python
# print all values that cannot be converted to float
for column_name in df.TOTAL_REVENUE:
    print("These are the problematic values for the variable: {}".format(column_name))
    for value in df.TOTAL_REVENUE:
        try:
            float(value)
        except:
            print(value)
```

```python
# 2 

dfcopy = df.copy()

fill_list = ["ENROLL", "TOTAL_REVENUE", "FEDERAL_REVENUE", 
             "STATE_REVENUE", "LOCAL_REVENUE", "TOTAL_EXPENDITURE", 
             "INSTRUCTION_EXPENDITURE", "SUPPORT_SERVICES_EXPENDITURE",
             "OTHER_EXPENDITURE", "CAPITAL_OUTLAY_EXPENDITURE", "GRADES_PK_G", 
             "GRADES_KG_G", "GRADES_4_G", "GRADES_8_G", "GRADES_12_G", "GRADES_1_8_G", 
             "GRADES_9_12_G", "GRADES_ALL_G"]
for col in fill_list:
    dfcopy.loc[:, col] = dfcopy.loc[:, col].fillna(dfcopy.loc[:, col].mean())

```

```python
# 3

dfcopy2 = df.copy()
years = dfcopy2['YEAR'].unique()

for col in fill_list:
    for year in years:
        dfcopy2.loc[dfcopy2['YEAR']==year,col] = dfcopy2.loc[dfcopy2['YEAR']==year, col].fillna(dfcopy2[dfcopy2['YEAR']==year][col].mean())
```

```python
# 4
dfcopy3 = df.copy()

for col in fill_list:
    dfcopy3.loc[:, col] = dfcopy3.loc[:, col].interpolate()
```

```python
for col in fill_list:
    print("Statistics for columns: {}".format(col))
    plswork = pd.concat([df[col], dfcopy[col], dfcopy2[col], dfcopy3[col]], axis=1).describe()
    print(plswork)
```

```python
fig = plt.figure(figsize=(20,100))

for i, col in enumerate(fill_list):
    df = plswork
    
    plt.subplot(len(fill_list), 4, i+1)
    plt.bar(["original", "overall", "year-by-year", "interpolation"], df.median())
    plt.xticks(rotation='vertical')
    plt.title("Mod of {}".format(col))
    
plt.tight_layout()
plt.show()
```

consistent pattern shown were interpolation returns the least amount, followed by the original, then year by year filter, and the the overall finally.
